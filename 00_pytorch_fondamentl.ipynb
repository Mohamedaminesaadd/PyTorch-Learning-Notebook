What is torch (PyTorch)?

The torch package in PyTorch is the core library used for deep learning.

It provides:

A data structure called tensors (multi-dimensional arrays)
Mathematical operations on tensors
Tools for GPU acceleration (CUDA)
Utilities for saving, loading, and managing models
Support for automatic differentiation (training neural networks)

Initializing a Tensor
Tensors can be initialized in various ways. Take a look at the following examples:

Directly from data

Tensors can be created directly from data. The data type is automatically inferred.

```python
data = [[1,2],[3,4]]
x_data = torch.tensor(data)
print(x_data)
```

From a NumPy array
Check type: is_tensor()
Create tensors: tensor(), zeros(), ones(), rand()
Convert data: from_numpy(), as_tensor()

PyTorch offers many ways to create tensors:
torch.zeros() → filled with 0
torch.ones() → filled with 1
torch.rand() → random values (0 to 1)
torch.randn() → random normal distribution
torch.arange() → sequence of numbers
torch.eye() → identity matrix

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(x_np)
```

create new tensors based on an existing tensor (x_data) in PyTorch.
Creates a new tensor filled with 1s, but with the same shape and properties as x_data

“retains the properties of x_data” means:
same shape (size)
same device (CPU or GPU)
same layout

```python
x_data = [[2, 5],
          [7, 9]]
x_data = torch.tensor(x_data)
x_ones = torch.ones_like(x_data)  # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")
```

reates a new tensor with:

same shape as x_data
but filled with random numbers (0 to 1)

```python
x_rand = torch.rand_like(x_data, dtype=torch.float)
print(f"a random Tensor: \n {x_rand} \n")
```

shape is a tuple of tensor dimensions. In the functions below, it determines the dimensionality of the output tensor.

```python
shape = (2,3)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)
print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")   
```

Arithmetic operations

```python
# return the transpose of tensor 
#@ is the matrix multiplication operator
y1 = tensor @ tensor.T
print(f"the result of matrix multiplication is \n {y1} \n")
y2 = tensor.matmul(tensor.T)
print(f"the result of matrix multiplication is \n {y2} \n")

#Element-wise multiplication
y3 = torch.rand_like(tensor)
z1 = y3 * tensor
print(f"the result of elementwise multiplication is \n {z1} \n")
z2 = y3.mul(tensor)
print(f"the result of elementwise multiplication is \n {z2} \n")    


#Do NOT create a new tensor — store the result directly in z3”
z3 = torch.rand_like(tensor)
torch.mul(tensor, z3, out=z3)
print(f"the result of elementwise multiplication is \n {z3} \n")

```

Single-element tensors If you have a one-element tensor, for example by aggregating all values of a tensor into one value, you can convert it to a Python numerical value using item():

```python
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

In PyTorch, in-place operations are operations that modify a tensor directly in memory instead of creating a new one

They are identified by a _ (underscore) at the end of the function name.

exemple 
copy_()
    x.copy_(y)
    copies values of y into x
    overwrites x


t_( ) transpose in-place
    x.t_()
    👉 Transposes x directly
    👉 No new tensor created


✔ Advantages:
saves memory
faster execution
useful for large tensors (deep learning)
❌ Risks:
can break gradient tracking
can overwrite needed values
harder to debug

```python
print(tensor, "\n")
tensor.add_(5)
print(tensor)
```

Bridge with NumPy

Tensor to NumPy array

```python
t = torch.ones(5)
print(f"{t} \n")
n = t.numpy()
print(f"{n} \n")
```

```python
t.add_(1) #change t in place
print(f"t: {t}")
print(f"n: {n}")

#A change in the tensor reflects in the NumPy array.
```

NumPy array to Tensor

```python
n = np.ones(5)
t = torch.from_numpy(n)
print(f"t: {t} \n")
print(f"n: {n} \n")
```

```python
#Changes in the NumPy array reflects in the tensor.
np.add(n, 1, out=n) #change in the numpy array
print(f"t: {t}")
print(f"n: {n}")
```
