Normalisation is one of the most important concepts in deep Learning .it helps neural networks train faster .more stable and often more accurately 
the probleme without normalisation 
suppose we have one feature is age and another is salary 
| Person | Age | Salary |
| ------ | --: | -----: |
| A      |  20 |  2,000 |
| B      |  25 |  5,000 |
| C      |  40 | 20,000 |*

the probleme is 
    Age 20 → 40
    Salary 2,000 → 20,000

large differnces like this make the optimisation harder 
normalisation transforms the values so they habe a similar scale

Batch normalisation :
    normalizes the output of a layer using the statistcs of the current mini-batch

    suppose the mini batch is     2  4  6  8
    step 1 :compute the mean m
    step 2 :compute the variance v
    step 3 : normalisation 
        Xn = (x-m)/squart(v+eps)

    now the data has approximately mean 0 variance = 1 
    step 4 : learnable scale and shift 
        batch normalisation does not stop there 
        it leans two parameters  γ(gamma) and β(beta )
        the fianle output is     y =γ*Xn+β
    this lets the network choose the best scale and offest instead of forcing every layer to have exactly zero mean and unit variance

Training vs Evaluation Mode 
batchNorm behaves differntly depending on whether the model is training or making predictions 

    During training 
    it computes the mean and varaince from the current mini-batch and also updates runing estimates 

    During infernce (evaluation)
    it uses the stored runing mean and runing variance colleted during training
    


Why is BatchNorm So Effective?

Imagine a runner trying to hit a moving target.

Without BatchNorm, the distribution of inputs to each layer changes as earlier layers update their weights. Each layer constantly has to adapt to new input distributions.

With BatchNorm, each layer receives inputs that are more stable:

Without BatchNorm

Layer 1 → Very large values
Layer 2 → Very small values
Layer 3 → Distribution keeps changing

Training becomes difficult.
With BatchNorm

Layer 1 → Normalized
Layer 2 → Normalized
Layer 3 → Normalized

Training becomes more stable and often converges faster.

there are a lot of normalisation layers ,but they normalize data in differnt ways.the biggest differnce is what they normalize over 

| Layer               | Normalizes Across       | Mostly Used In           |
| ------------------- | ----------------------- | ------------------------ |
| `BatchNorm`         | Batch                   | CNNs                     |
| `GroupNorm`         | Groups of channels      | CNNs (small batch sizes) |
| `InstanceNorm`      | One sample, one channel | Style transfer           |
| `LayerNorm`         | Features of one sample  | Transformers, LLMs       |
| `RMSNorm`           | RMS of one sample       | Modern LLMs              |
| `SyncBatchNorm`     | Multiple GPUs           | Distributed training     |
| `LocalResponseNorm` | Neighboring channels    | Old CNNs (AlexNet)       |



1.GroupNorm
was instroduced because batch normalise perfomrs prooly when the batch size is small 
suppose you have a tensoe 
    Shape
    (2,8,32,32)
    Batch = 2
    Channels = 8

batchNorm computes statics over the batch 
but what if the batch is = 1 
The statistics become unreliable.
GroupNorm solves this by splitting the channels into groups.
    example channels = 8 , groups =4 
    12|34|56|78
    Each group is normalized independently.

it'iq used when the number of the batch is so small 


```python
import torch
import torch.nn as nn 

x= torch.randn(2,8,32,32)

gn = nn.GroupNorm(
    num_groups=4,num_channels=8
)
y= gn(x)
print(y.shape)
```

visualisation  it simulate a lot of batch and perfome the normal normalization 


2.nn.syncbatchNorm
suppose we have train on 4 GPUs
gpu1: batch = 8
gpu2: batch = 8
gpu3: batch = 8
gpu4: batch = 8

normal BatchNorm computes 
GPU1 Mean
GPU2 Mean
GPU3 Mean
GPU4 Mean
    each GPU has differnt statistics

SyncBatchNorm synchronizes all GPUs.
instead 
    GPU1
    ↓
    GPU2
    ↓
    GPU3
    ↓
    GPU4
    ↓
    Shared Mean
    ↓
    Shared Variance

all the GPUs use exactly the same normalisation 

Applications
    Distributed Training
    Multi-GPU Training
    Used in
    Large CNNs
    Large Vision Models

4.nn.LayerNorm
layerNorm is one of the most important normalisation methodes today 
used in 
    GPT 
    BERT 
    Llama
    VISION TRANSFOMER

Instead of normalizing across the batch ,
layerNorm normalizes each sample independently across its features 
    suppose one sample [3,5,8,2]
    compute variance using only these four numbers 
    then normalize them 
    the next sample is normalized independently 


```python
import torch
import torch.nn as nn

x = torch.randn(32,512)

ln = nn.LayerNorm(512)

y = ln(x)
```

visualization

    Sample 2

    ↓

no dependence on the batch 

Why is this useful?
Transformers often use Batch = 1
    during inference.
    BatchNorm would fail.
    LayerNorm works perfectly.


| Layer               | Normalizes Across                         | Depends on Batch? | Main Applications                                   |
| ------------------- | ----------------------------------------- | ----------------- | --------------------------------------------------- |
| `BatchNorm`         | Batch + channels                          | ✅ Yes             | CNNs with moderate or large batch sizes             |
| `GroupNorm`         | Groups of channels                        | ❌ No              | Detection, segmentation, small-batch CNNs           |
| `SyncBatchNorm`     | Batch across all GPUs                     | ✅ Yes             | Multi-GPU distributed training                      |
| `InstanceNorm`      | Each sample, each channel                 | ❌ No              | Style transfer, GANs                                |
| `LayerNorm`         | Features of one sample                    | ❌ No              | Transformers, NLP, LLMs                             |
| `LocalResponseNorm` | Neighboring channels                      | ❌ No              | Older CNNs like AlexNet                             |
| `RMSNorm`           | Root mean square of one sample's features | ❌ No              | Modern LLMs (Llama, Mistral, Qwen)                  |
| `LazyInstanceNorm*` | Same as InstanceNorm                      | ❌ No              | When the number of channels is not known in advance |
