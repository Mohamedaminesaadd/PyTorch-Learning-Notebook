maxPooling
Max Pooling is a layer used in CNNs to reduce the size of feature maps while keeping the most important information.

MaxPool1d

```python
import torch
import torch.nn as nn

x = torch.arange(1,17).float().reshape(1,1,16)

pool = nn.MaxPool1d(kernel_size=2)

y = pool(x)

print(y)
```

Adaptive Pooling Layers

These automatically choose the pooling region size to produce a desired output size.

```python
import torch 
import torch.nn as nn 
x = torch.arange(1,17).float().reshape(1,1,4,4)

pool2d = nn.AdaptiveAvgPool2d(2)
y = pool2d(x)

print(x ,"\n")
print(y,"\n")
print(y.shape)



```

Fractionnal max avergre 1D 2D 3D 

1.nn.FractionalMaxPool2d
Normal maxpooling uses a fixed stride 
exemple nn.MaxPool2d(kernel_size=2, stride=2)

    4×4 → 2×2
    8×8 → 4×4
    16×16 → 8×8

the reduction factor is always fixed

but with FractionalMaxPool2d ,the reduction factor can be non-integer
exemple :32×32 → 20×20
instead od :32×32 → 16×16 or 32×32 → 8×8


```python
import torch
import torch.nn as nn

x = torch.randn(1, 1, 8, 8)

pool = nn.FractionalMaxPool2d(
    kernel_size=2,
    output_size=(5,5)
)

y = pool(x)

print(y.shape)
```

Allows more flexibility in the output size.

Used in some research papers to add randomness and reduce overfitting.

2. nn.FractionalMaxPool3d
same idea but for 3D data 
input (N,C,D,H,W)


```python
import torch
import torch.nn as nn

x = torch.randn(1, 1, 8, 8,8)


pool = nn.FractionalMaxPool3d(
    kernel_size=2,
    output_size=(5,5,5)
)

y = pool(x)

print(y.shape)
```

LPPooling
LP Pooling is based on the Lp norm.
formula (somme de xi ^p)^1/p   
p norm degre 


```python
import torch
import torch.nn as nn

x = torch.tensor([[[3.,4.,5.,6.]]])

pool = nn.LPPool1d(
    norm_type=2,
    kernel_size=2
)
# for nrom_type= 2 is like a averge pooling 
#for nrom_type= infii  is like a max pooling 

y = pool(x)

print(y)
```

LP pooling is a compromise between
Average Pooling
      and
Max Pooling

p = 1  → close to Average Pooling
p = 2  → Euclidean norm
p = 10 → very close to Max Pooling
p → ∞  → Max Pooling

and is the same of 2D et 3D 

| Layer                 | Input Type     | Purpose                     |
| --------------------- | -------------- | --------------------------- |
| `FractionalMaxPool2d` | Images         | Flexible/random max pooling |
| `FractionalMaxPool3d` | Videos/Volumes | Flexible 3D max pooling     |
| `LPPool1d`            | Signals        | Lp-norm pooling             |
| `LPPool2d`            | Images         | Lp-norm pooling on images   |
| `LPPool3d`            | Volumes/Videos | Lp-norm pooling on 3D data  |


Average Pooling:
"Take the average"

Max Pooling:
"Take the strongest value"

LP Pooling:
"Take a value between average and max,
depending on p"

Fractional Max Pooling:
"Max Pooling with flexible output size"
