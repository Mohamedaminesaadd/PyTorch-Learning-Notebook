Fractionnal max avergre 1D 2D 3D 

1.nn.FractionalMaxPool2d
Normal maxpooling uses a fixed stride 
exemple nn.MaxPool2d(kernel_size=2, stride=2)

    4×4 → 2×2
    8×8 → 4×4
    16×16 → 8×8

the reduction factor is always fixed

but with FractionalMaxPool2d ,the reduction factor can be non-integer
exemple :32×32 → 20×20
instead od :32×32 → 16×16 or 32×32 → 8×8


```python
import torch
import torch.nn as nn

x = torch.randn(1, 1, 8, 8)

pool = nn.FractionalMaxPool2d(
    kernel_size=2,
    output_size=(5,5)
)

y = pool(x)

print(y.shape)
```

Allows more flexibility in the output size.

Used in some research papers to add randomness and reduce overfitting.

2. nn.FractionalMaxPool3d
same idea but for 3D data 
input (N,C,D,H,W)


```python
import torch
import torch.nn as nn

x = torch.randn(1, 1, 8, 8,8)


pool = nn.FractionalMaxPool3d(
    kernel_size=2,
    output_size=(5,5,5)
)

y = pool(x)

print(y.shape)
```

what is the differnt between the adaptiveMaxPool ans FractionalMaxPool

=>Although FractionalMaxPool and AdaptiveMaxPool both reduce the size of a feature map using max pooling, they work in completely different ways.

1.AdaptiveMaxpool 
    You specify the output size, and PyTorch `automatically` chooses the kernel size and stride.
    You don't care about the pooling window.

=>why use it :if you have input a differnt image size and you want a one output value like (3*3)
This is why almost every CNN classifier uses Adaptive Pooling before the fully connected layer.

2.FractionalMaxPool 
    Fractional pooling is very different.
    Instead of choosing a fixed kernel size,
    it chooses non-uniform pooling regions.
    The pooling windows have different sizes.

    for exemple 
    instead of 
    2×2
    2×2
    2×2
    2×2

    it may choose 
    2×2
    3×2
    2×3
    3×3

    The regions are chosen randomly (or pseudo-randomly).
    suppose the input is 8x8
    and you want approxmately 5x5
    a normal maxpool cannot do this nicely because 8/5=1.6
    there is no kernel size 1.6 × 1.6 to give us  5x5 outpout images
    =>Fractional pooling solves this problem.

                Why use FractionalMaxPool?
                It was introduced in the paper:
                Fractional Max-Pooling (Benjamin Graham, 2014)
                The idea was to:
                reduce overfitting,
                introduce randomness similar to dropout,
                improve generalization,
                allow smoother reductions in feature map size.
                Instead of always shrinking by exactly 2,
                64 → 32 → 16 → 8
                you can shrink more gradually,
                64 → 45 → 32 → 22 → 16
                This preserves more information.

| Feature                  | AdaptiveMaxPool                                  | FractionalMaxPool                                        |
| ------------------------ | ------------------------------------------------ | -------------------------------------------------------- |
| Goal                     | Produce a specific output size                   | Produce a smaller output using irregular pooling regions |
| Pooling regions          | Computed deterministically                       | Chosen pseudo-randomly                                   |
| Randomness               | No                                               | Yes                                                      |
| Output size              | Exact                                            | Exact (`output_size`) or approximate (`output_ratio`)    |
| Kernel size              | Computed implicitly from the desired output      | Specified by the user (e.g., `kernel_size=2`)            |
| Typical use              | Standard CNNs, classifiers, variable-size inputs | Research, regularization, gradual downsampling           |
| Common in modern models? | **Very common**                                  | **Rare**                                                 |

                

```python
import torch 
import torch.nn as nn 
x = torch.arange(1,17).float().reshape(1,1,4,4)

pool2d = nn.AdaptiveMaxPool2d((3,1))
y= pool2d(x)

print(x ,"\n")
print(y,"\n")
print(y.shape)
```

a Padding Layers adds extra values arouand the border of the input tensor before appling operations like convolution or pooling 
padding is mainly used to:
1.preserve the spatial size of the input .
2.prevent losing information at the bordes .
3.control the output size of convolution layers 

why is padding needed 
consider a 3*3 convolution kernel 
without padding :
    Input: 5×5
    Kernel: 3×3
    Stride: 1

output size :
        (5-3)/1 + 1 = 3 
        so 5*5 ====> 3*3
        problem the image becomes smaller

with padding =1 : the output is 5*5 

nn.ReflectionPad1d
pads a 1D tensor by reflecting mirroring the values at the boundary 
suppose we have 1234
the result is 
32|1234|32
padding = 2 


```python
import torch 
import torch.nn as nn 

x = torch.tensor([1,2,3,4] ,dtype=float)
pad = nn.ReflectionPad1d(2)
y = pad(x.unsqueeze(0))
print(y)
```

let's applied for 2D 
the same idea for the images (tensor 2D)

```python
import torch
import torch.nn as nn

x = torch.randint(9, (1,1,3,3))
#x = torch.randint(low=0, high=9, size=(1,1,3,3))

pad2d = nn.ReflectionPad2d(1)
y = pad2d(x)

print(x)
print()
print(y)
```

The border is created by mirroring the interior values, not by repeating the edge values.
